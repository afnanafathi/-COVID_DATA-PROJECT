```python
import requests
import csv
```

```python
response = requests.get("http://coronavirus.m.pipedream.net/")
response
```

```python
covid = response.json()
covid
```

```python
covid['rawData']
```

```python
with open('COVID_Data.csv','w',newline= '') as f:
    writer= csv.DictWriter(f,fieldnames=['Country_name','Province_State','Confirmed','Deaths'])
    writer.writeheader()

    for c in covid['rawData']:
        country=c['Country_Region']
        province=c['Province_State']
        confirmed= c['Confirmed']
        deaths= c['Deaths']

        writer.writerow(
            {
                'Country_name':  country,
                'Province_State':  province,
                'Confirmed':  confirmed,
                'Deaths':  deaths
            }
        )
```
